```python
#7. Generate mark sheet with grades

eng = 40
maths = 60
urdu = 80
computer =89
science = 50


total = eng+maths+urdu+computer+science
   
    

percentage = total / 5

print("Total Marks:", total)
print("Percentage:", percentage)

if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

print("Grade:", grade)
```
