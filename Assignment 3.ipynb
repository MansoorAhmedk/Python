```python
##Python Calculator (Addition, Subtraction, Multiplication, Division, Power)


num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)

if num2 != 0:
    print("Division:", num1 / num2)
else:
    print("Division: Cannot divide by zero")

print("Power:", num1 ** num2)
```

```python
#Check if There is Any Numeric Value in a List (Using for loop)
my_list = ["apple", "banana", 10, "orange"]

found = False

for item in my_list:
    if isinstance(item, (int, float)):
        found = True
        break

if found:
    print("Numeric value exists in the list.")
else:
    print("No numeric value found.")
```

```python
#Add a Key to a Dictionary
my_dict = {
    "name": "Ali",
    "age": 25
}

# Adding new key
my_dict["city"] = "Karachi"

print(my_dict)
```

```python
#Sum All Numeric Items in a Dictionary
my_dict = {
    "a": 10,
    "b": 20,
    "c": 30
}

total = sum(my_dict.values())

print("Sum of numeric values:", total)
```

```python
#Identify Duplicate Values in a List
my_list = [1, 2, 3, 4, 2, 5, 1, 6]

duplicates = []

for item in my_list:
    if my_list.count(item) > 1 and item not in duplicates:
        duplicates.append(item)

print("Duplicate values:", duplicates)
```

```python
#Check if a Key Exists in a Dictionary
my_dict = {
    "name": "Ahmed",
    "age": 30,
    "city": "Lahore"
}

key = input("Enter key to check: ")

if key in my_dict:
    print("Key exists in dictionary.")
else:
    print("Key does not exist.")
```
